# SLURM Magics Quickstart\n\nInteract with a remote SLURM cluster over SSH using IPython magics. This example shows configuration via environment variables or %slurm_config, querying the queue, and submitting a simple job.

```python
%load_ext slurm.magics
```

If the above fails, try the fully-qualified module path: `%load_ext ads.slurm.magics`.

```python
%slurm_help
```

## Configure (Option A): environment variables

```python
import os\nos.environ["SLURM_SSH_HOST"] = "user@login-node"\nos.environ["SLURM_SSH_KEY"] = os.path.expanduser("~/.ssh/id_rsa")\nos.environ["SLURM_CONF_REMOTE"] = "/etc/slurm/slurm.conf"\n# Optional:\n# os.environ["SLURM_SSH_JUMP"] = "user@bastion"
```

## Configure (Option B): %slurm_config

```python
%slurm_config --host user@login-node --key ~/.ssh/id_rsa --conf /etc/slurm/slurm.conf --jump user@bastion
```
